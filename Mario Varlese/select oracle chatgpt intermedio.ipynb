```python
%%timeit -r 10 -n 30

import cx_Oracle
from datetime import datetime
import json

def query_oracle_db():
    # Parametri di connessione al database Oracle
    db_params = {
        "hostname": "localhost",
        "user": "SYSTEM",
        "password": "oirad1993",
        "service_name": "XEPDB1"
    }

    try:
        # Connessione al database
        connection = cx_Oracle.connect(
            user=db_params["user"],
            password=db_params["password"],
            dsn=f"{db_params['hostname']}/{db_params['service_name']}"
        )

        # Creazione di un cursore
        cursor = connection.cursor()

        # Esecuzione della query
        query = """
            SELECT *
            FROM "APPARECCHIATUREELETTRONICHE"
            WHERE TO_NUMBER(SUBSTR("RISOLUZIONESCHERMO", INSTR("RISOLUZIONESCHERMO", 'x')+1)) > 1080
            ORDER BY "DATAPRODUZIONE" DESC
        """

        cursor.execute(query)

        # Recupero dei risultati
        result = cursor.fetchall()

        if not result:
            # Nessun risultato trovato
            response = {'status_code': 404, 'subcode': 5}
        else:
            # Risultato trovato
            current_timestamp = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
            response = {'data': result, 'request_timestamp': str(current_timestamp)}

        return response

    except cx_Oracle.DatabaseError as e:
        # Gestione delle eccezioni del database
        error_message = str(e)
        print(f"Errore del database: {error_message}")
        raise

    except Exception as e:
        # Gestione di altre eccezioni
        print(f"Errore generico: {str(e)}")
        raise

    finally:
        # Chiusura del cursore e della connessione nel blocco finally
        if cursor:
            cursor.close()
        if connection:
            connection.close()

# Esempio di utilizzo della funzione
try:
    result = query_oracle_db()
    print(json.dumps(result, indent=2))

except Exception as e:
    # Gestione delle eccezioni globali
    print(f"Errore: {str(e)}")
```

```python
%%timeit -r 10 -n 30

'''
Conversione query ricerca Apparecchiature Elettroniche in python
SELECT *
FROM "apparecchiatureelettroniche"
WHERE CAST(SPLIT_PART("risoluzioneschermo", 'x', 2) AS INT) > 1080
ORDER BY "dataproduzione" DESC;

Vorrei mi traducessi la query indicata utilizzando il linguaggio python, tieni presente che il backend di riferimento utilizzera un database PostgressSql
Utilizza la libreria psycopg2
Evidenziami i parametri che potrebbero essere sucettibili di Sql-Injection e proponimi una versione sicura della query che utilizzi i prepared statements.
#import delle librerie
#definizione di variabili
#funzione per connettersi al database 
#funzione per eseguire la query 
#funzione per stampare i risultati 
#main

'''

import psycopg2
import pandas as pd

#definizione di variabili
host = "localhost"
user= "postgres"
database = "database4test"
password = "oirad1993"

#funzione per connettersi al database
def connect_db(host, database, password, user):
    try:
        conn = psycopg2.connect(host=host, database=database, user=user, password=password)
        print("Connected to the database")
        return conn
    except Exception as e:
        print("Unable to connect to the database")
        raise e

#funzione per eseguire la query
def execute_query(conn, query):
    try:
        cur = conn.cursor()
        cur.execute(query)
        print("Query executed successfully")
        return cur
    except:
        print("Unable to execute the query")

#funzione per stampare i risultati
def print_results(cur):
    try:
        rows = cur.fetchall()
        df = pd.DataFrame(rows, columns=['id', 'marca', 'modello', 'dataproduzione', 'risoluzioneschermo'])
        print(df)
    except:
        print("Unable to print the results")

#main
conn = connect_db(host, database, password, user)
query = '''SELECT *
FROM "apparecchiatureelettroniche"
WHERE CAST(SPLIT_PART("risoluzioneschermo", 'x', 2) AS INT) > 1080
ORDER BY "dataproduzione" DESC;'''
cur = execute_query(conn, query)
print_results(cur)
conn.close()

# Path: Untitled-1.ipynb
```

```python
%%timeit -r 10 -n 30
import mysql.connector

def main():
    # Connessione al database
    connection = mysql.connector.connect(
        host="localhost",
        user="root",
        password="oirad1993",
        database="database4test"
    )

    # Creazione del cursore
    cursor = connection.cursor()

    # Eseguire la query
    cursor.execute("""
        SELECT
            ae.TipoProdotto,
            YEAR(ae.DataProduzione) AS AnnoProduzione,
            SUM(LENGTH(ae.ManualeUtente)) AS DimensioneTotaleManuali
        FROM ApparecchiatureElettroniche ae
        GROUP BY ae.TipoProdotto, YEAR(ae.DataProduzione)
        ORDER BY ae.TipoProdotto, YEAR(ae.DataProduzione);
    """)

    # Ottenere i risultati
    risultati = cursor.fetchall()

    # Stampare i risultati
    for risultato in risultati:
        print(risultato)

    # Chiudere la connessione
    connection.close()

if __name__ == "__main__":
    main()
```

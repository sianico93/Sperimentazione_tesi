```python
%%timeit -r 10 -n 30
import psycopg2
from psycopg2 import sql

# Connessione al database
conn = psycopg2.connect(
    dbname="database4test",
    user="postgres",
    password="oirad1993",
    host="localhost",
    port="5432"
)

# Creazione di un cursore per eseguire le query
cur = conn.cursor()

# Esecuzione della query
query = sql.SQL("""
    SELECT 
        ae.tipoprodotto,
        EXTRACT(YEAR FROM ae.dataproduzione) AS annoproduzione,
        SUM(OCTET_LENGTH(ae.manualeutente)) AS dimensionetotalemanuali
    FROM apparecchiatureelettroniche ae
    GROUP BY ae.tipoprodotto, EXTRACT(YEAR FROM ae.dataproduzione);
""")

cur.execute(query)

# Recupero dei risultati
risultati = cur.fetchall()

# Chiusura della connessione
cur.close()
conn.close()

# Stampa dei risultati
for riga in risultati:
    tipoprodotto, annoproduzione, dimensionetotalemanuali = riga
    print(f"Tipo Prodotto: {tipoprodotto}, Anno Produzione: {annoproduzione}, Dimensione Totale Manuali: {dimensionetotalemanuali}")
```

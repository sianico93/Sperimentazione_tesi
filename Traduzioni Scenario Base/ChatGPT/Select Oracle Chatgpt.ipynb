```python
%%timeit -r 10 -n 30

import cx_Oracle

# Connect to Oracle database
connection = cx_Oracle.connect('SYSTEM', 'oirad1993', 'localhost:1521/XEODB1')

# Create a cursor
cursor = connection.cursor()

# SQL query
sql_query = """
    SELECT *
    FROM APPARECCHIATUREELETTRONICHE
    WHERE TO_NUMBER(SUBSTR(RISOLUZIONESCHERMO, INSTR(RISOLUZIONESCHERMO, 'x')+1)) > 1080
    ORDER BY DATAPRODUZIONE DESC
"""

# Execute the query
cursor.execute(sql_query)

# Fetch all rows
result_set = cursor.fetchall()

# Print the results
for row in result_set:
    print(row)

# Close cursor and connection
cursor.close()
connection.close()
```

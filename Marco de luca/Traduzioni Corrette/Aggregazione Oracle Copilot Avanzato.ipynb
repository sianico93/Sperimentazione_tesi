```python
import os
import cx_Oracle
import pandas as pd

def connect():
    try:
        connection = cx_Oracle.connect("SYSTEM", "oirad1993", "localhost:1521/XEPDB1")
        return connection
    except cx_Oracle.DatabaseError as e:
        print("Errore di connessione al database: ", e)
        exit(1)

def disconnect(connection):
    try:
        connection.close()
    except cx_Oracle.DatabaseError as e:
        print("Errore di disconnessione al database: ", e)
        exit(1)

def query(connection):
    try:
        cursor = connection.cursor()
        cursor.execute("SELECT AE.TIPOPRODOTTO, TO_CHAR(AE.DATAPRODUZIONE, 'YYYY') AS ANNOPRODUZIONE, SUM(DBMS_LOB.GETLENGTH(AE.MANUALEUTENTE)) AS DIMENSIONETOTALEMANUALI FROM APPARECCHIATUREELETTRONICHE AE GROUP BY AE.TIPOPRODOTTO, TO_CHAR(AE.DATAPRODUZIONE, 'YYYY')")
        return cursor
    except cx_Oracle.DatabaseError as e:
        print("Errore di esecuzione della query: ", e)
        exit(1)

#codice
connection = connect()
cursor = query(connection)
df = pd.DataFrame(cursor, columns=['TIPOPRODOTTO', 'ANNOPRODUZIONE', 'DIMENSIONETOTALEMANUALI'])
print(df)
disconnect(connection)
```

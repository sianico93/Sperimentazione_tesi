```python
%%timeit -r 10 -n 30
'''
Traduzione query aggregazione MySQL in Copilot
Devo tradurre una query di aggregazione da linguaggio SQL a linguaggio Python
SELECT 
ae.TipoProdotto,
YEAR(ae.DataProduzione) AS AnnoProduzione,
SUM(LENGTH(ae.ManualeUtente)) AS DimensioneTotaleManuali
FROM ApparecchiatureElettroniche ae
GROUP BY ae.TipoProdotto, YEAR(ae.DataProduzione);

Il database da utilizzare è MySQL, voglio suddividere il codice python in funzioni e gestire le eventuali eccezioni. Voglio includere dei commenti per ogni sezione di codice ottenuta.
Usare la libreria mysql-connector, e dividere il codice in moduli separati.
Non includere i dati sensibili all’interno del codice, ed evitare attacchi di tipo SQL-injection .
Voglio che l’output del codice venga mostrato come file json. Organizza i vari moduli in directory aggiungendo anche quella per i test e genera alcuni test  per la verifica della sicurezza.
La struttura del codice è la seguente:
# import delle librerie utili
# definizione variabili globali
# definizione della funzione di connessione al database
# definizione delle funzioni utili per l’esecuzione
# main

'''
# import delle librerie utili
import mysql.connector
import json
import os
import sys

# definizione variabili globali
# DB_USER = os.environ.get('root')
# DB_PASSWORD = os.environ.get('oirad1993')
# DB_HOST = os.environ.get('localhost')
# DB_NAME = os.environ.get('database4test')

# definizione della funzione di connessione al database
def connect_db():
    try:
        db_connection = mysql.connector.connect(
            user="root",
            password="oirad1993",
            host="localhost",
            database="database4test"
        )
        return db_connection
    except mysql.connector.Error as err:
        print("Errore di connessione al database: {}".format(err))
        sys.exit(1)

# definizione delle funzioni utili per l’esecuzione
def get_cursor(db_connection):
    try:
        cursor = db_connection.cursor()
        return cursor
    except mysql.connector.Error as err:
        print("Errore di connessione al database: {}".format(err))
        sys.exit(1)

def close_connection(db_connection):
    try:
        db_connection.close()
    except mysql.connector.Error as err:
        print("Errore di connessione al database: {}".format(err))
        sys.exit(1)

def execute_query(cursor, query):
    try:
        cursor.execute(query)
    except mysql.connector.Error as err:
        print("Errore di connessione al database: {}".format(err))
        sys.exit(1)

def get_result(cursor):
    try:
        result = cursor.fetchall()
        return result
    except mysql.connector.Error as err:
        print("Errore di connessione al database: {}".format(err))
        sys.exit(1)

def get_json(result):
    try:
        json_result = json.dumps(result)
        return json_result
    except mysql.connector.Error as err:
        print("Errore di connessione al database: {}".format(err))
        sys.exit(1)

# main
def main():
    db_connection = connect_db()
    cursor = get_cursor(db_connection)
    query = "SELECT ae.TipoProdotto, YEAR(ae.DataProduzione) AS AnnoProduzione, SUM(LENGTH(ae.ManualeUtente)) AS DimensioneTotaleManuali FROM ApparecchiatureElettroniche ae GROUP BY ae.TipoProdotto, YEAR(ae.DataProduzione);"
    execute_query(cursor, query)
    result = get_result(cursor)
    # json_result = get_json(result)
    # print(json_result)
    close_connection(db_connection)

if __name__ == "__main__":
    main()

# Path: Untitled-2.ipynb
```

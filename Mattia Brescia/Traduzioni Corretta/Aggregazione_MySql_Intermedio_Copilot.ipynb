```python
%%timeit -r 10 -n 30

'''
Traduci la seguente query aggregata da linguaggio sql a Python per il database che ho creato MySql.

SELECT 
ae.TipoProdotto,
YEAR(ae.DataProduzione) AS AnnoProduzione,
SUM(LENGTH(ae.ManualeUtente)) AS DimensioneTotaleManuali
FROM ApparecchiatureElettroniche ae
GROUP BY ae.TipoProdotto, YEAR(ae.DataProduzione);

La query deve essere strutturata con funzioni. Gestisci le eccezioni che si possono verificare trattandole opportunamente e segnalandole a chi invoca la query. 
Commenta il codice utilizzando parole chiave in modo che sia leggibile da terzi.
Deve prevedere utilizzo delle opportune librerie come MySql Connector
Il codice che devi produrre deve consentirmi di accedere al DB evitando di creare eventuali vulnerabilit che possono essere sfruttate per attacchi. Poni particolare attenzione ai dati sensibili evitando che questi vengano sovrascritti.
Mostra l'output in una struttura tabellare.

#import librerie
#funzione di connessione al database
#funzione di disconnessione al database
#funzione di esecuzione della query
#main in cui vengono chiamate le funzioni definite
#esecuzione del main e stampa degli output

'''


import mysql.connector
from mysql.connector import Error
import pandas as pd
import numpy as np

#funzione di connessione al database
def connect():
    """ Connect to MySQL database """
    try:
        conn = mysql.connector.connect(host='localhost',
                                       database='database4test',
                                       user='root',
                                       password='oirad1993')
        if conn.is_connected():
            print('Connected to MySQL database')
            return conn

    except Error as e:
        print(e)

#funzione di disconnessione al database
def disconnect(conn):
    """ Disconnect from MySQL database """
    try:
        conn.close()
        print('Disconnected from MySQL database')
    except Error as e:
        print(e)

#funzione di esecuzione della query
def execute_query(conn, query):
    """ Execute SQL query """
    try:
        cursor = conn.cursor()
        cursor.execute(query)
        print("Query executed successfully")
        return cursor

    except Error as e:
        print(e)

#main in cui vengono chiamate le funzioni definite
def main():
    conn = connect()
    query = "SELECT ae.TipoProdotto, YEAR(ae.DataProduzione) AS AnnoProduzione, SUM(LENGTH(ae.ManualeUtente)) AS DimensioneTotaleManuali FROM ApparecchiatureElettroniche ae GROUP BY ae.TipoProdotto, YEAR(ae.DataProduzione)"
    cursor = execute_query(conn, query)
    df = pd.DataFrame(cursor.fetchall())
    df.columns = cursor.column_names
    print(df)

#esecuzione del main e stampa degli output
if __name__ == '__main__':
    main()
```
